```python
def pdf_merge(folder, _file, _numstr):
    from PyPDF2 import PdfFileReader, PdfFileMerger
    _files = [os.path.join(folder,f) for f in os.listdir(folder) if '.pdf' in f and _numstr in f]
    # Merge the files
    merger = PdfFileMerger()
    for f in _files:
        merger.append(PdfFileReader(f), 'rb')
    merger.write(_file)
```
